```python
cell_type = {'excitatory': ['PC', 'SS', 'SP'],
             'inhibitory': ['AC', 'BP', 'BC', 'BTC', 'ChC', 'DBC', 'MC', 'NGC']}

for cell_type_name, cell_list in cell_type.items():
    for cell in cell_list:
        template_params['n'] = 50
        template_params['probe'] = 'Neuronexus-32'
        tempgen = mr.gen_templates(cell_models_folder=cell_folder, params=template_params, n_jobs=2)
        mr.save_template_generator(tempgen, filename=f'/media/ubuntu/sda/Spike_Sorting/paper_architecture/02_simulation_data/01_Neuroxenus_32_channels/data_generation/setting_6_neuron_type/Neuronexus_32_50_cell_cell_type_{cell}_templates.h5')
        templates_filename = f'/media/ubuntu/sda/Spike_Sorting/paper_architecture/02_simulation_data/01_Neuroxenus_32_channels/data_generation/setting_6_neuron_type/Neuronexus_32_50_cell_cell_type_{cell}_templates.h5'
        #tempgen = mr.load_templates(templates_filename)

        recording_params = mr.get_default_recordings_params()
        if cell_type_name == 'excitatory':
            recording_params['spiketrains']['n_exc'] = 50
            recording_params['spiketrains']['n_inh'] = 0
            recording_params['cell_types'] = {'excitatory': [cell], 'inhibitory': []}
        else:
            recording_params['spiketrains']['n_exc'] = 0
            recording_params['spiketrains']['n_inh'] = 50
            recording_params['cell_types'] = {'excitatory': [], 'inhibitory': [cell]}

        recording_params['spiketrains']['duration'] = 1800
        recording_params['spiketrains']['seed'] = 1


        recording_params['templates']['min_amp'] = 40
        recording_params['templates']['max_amp'] = 300
        recording_params['templates']['seed'] = 1
        recording_params['templates']['overlap_threshold'] = 0.9
        recording_params['templates']['min_dist'] = 25
        

        recording_params['recordings']['modulation'] = 'electrode'
        recording_params['recordings']['noise_mode'] = 'uncorrelated'
        recording_params['recordings']['noise_level'] = noise_level

        recording_params['recordings']['chunk_conv_duration'] = 20
        recording_params['recordings']['chunk_noise_duration'] = 20
        recording_params['recordings']['chunk_filter_duration'] = 20
        recording_params['recordings']['seed'] = 1
        recording_params['recordings']['fs'] = 10000

        with open(f'/media/ubuntu/sda/Spike_Sorting/paper_architecture/02_simulation_data/01_Neuroxenus_32_channels/data_generation/setting_6_neuron_type/Neuronexus_32_50_cell_cell_type_{cell}_recording_param.pkl', 'wb') as f:
            pickle.dump(recording_params, f)

        recgen = mr.gen_recordings(templates=templates_filename, params=recording_params, verbose=True)
        mr.save_recording_generator(recgen, f'/media/ubuntu/sda/Spike_Sorting/paper_architecture/02_simulation_data/01_Neuroxenus_32_channels/data_generation/setting_6_neuron_type/Neuronexus_32_50_cell_cell_type_{cell}_recording.h5')
```
