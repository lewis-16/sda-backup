```python
from PIL import Image
import os

def open_and_convert_image(image_folderpath, image_label):
    # 构造可能的文件路径
    jpg_path = os.path.join(image_folderpath, f"{image_label}.jpg")
    png_path = os.path.join(image_folderpath, f"{image_label}.png")
    
    # 尝试打开 .png 文件
    if os.path.exists(png_path):
        image_path = png_path
    # 如果没有 .png 文件，尝试打开 .jpg 文件
    elif os.path.exists(jpg_path):
        image_path = jpg_path
    else:
        print(f"No image found for label {image_label}")
        return None
    
    # 打开图像
    image = Image.open(image_path)
    
    # 如果图像是 PNG 格式且包含透明通道，保留透明通道
    if image.mode == 'RGBA':
        # 分离 RGBA 通道
        r, g, b, a = image.split()
        # 将 RGB 通道转换为灰度
        gray = Image.merge("RGB", (r, g, b)).convert("L")
        # 将灰度图像与透明通道重新合并
        gray_with_alpha = Image.merge("LA", (gray, a))
        return gray_with_alpha
    else:
        # 如果图像是 JPG 格式，直接转换为灰度模式
        return image.convert("L")

image_folderpath = "/root/visual_decode/nat_image_mask"  

for i in range(1, 118):
    image_label = str(i)
    converted_image = open_and_convert_image(image_folderpath, image_label)

    if converted_image:
        output_path = os.path.join("/root/visual_decode/nat_image_mask_new", f"{image_label}.png")
        converted_image.save(output_path)
        print(f"Converted image saved to {output_path}")
```
