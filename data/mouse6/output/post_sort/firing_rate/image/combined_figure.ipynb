```python
from PIL import Image
import os

entries =  os.listdir('/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image')
folders = [entry for entry in entries if os.path.isdir(os.path.join('/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image', entry))]
for folder in folders:
    image_folder = '/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image/'+folder
    image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

    image_files.sort(key=lambda x: int(x.split('_')[1].split('.')[0]))

    if len(image_files) != 117:
        raise ValueError("图片数量不正确，应有117张图片。")

    images = [Image.open(os.path.join(image_folder, img)) for img in image_files]

    image_width, image_height = images[0].size

    for img in images:
        img = img.resize((image_width, image_height))

    combined_image = Image.new('RGB', (13 * image_width, 9 * image_height))

    for i, img in enumerate(images):
        row = i // 13
        col = i % 13
        combined_image.paste(img, (col * image_width, row * image_height))

    combined_image.save(f'{folder}.jpg', quality=50) 
    combined_image.show()
```

```python
entries =  os.listdir('/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image')
folders = [entry for entry in entries if os.path.isdir(os.path.join('/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image', entry))]

os.mkdir("/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image/combined_image")
image_count = 117

for image_index in range(1, image_count + 1):
    image_path = os.path.join('/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image', folders[0], f"Image_{image_index}.png")
    first_image = Image.open(image_path)
    image_width, image_height = first_image.size
    combined_image = Image.new('RGB', (4 * image_width, 4 * image_height))

    for i, folder in enumerate(folders):
        row = i // 4
        col = i % 4
        image_path = os.path.join('/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image', folder, f"Image_{image_index}.png")
        img = Image.open(image_path)
        combined_image.paste(img, (col * image_width, row * image_height))

    
    output_path = f"/media/ubuntu/sda/data/mouse6/output/post_sort/firing_rate/image/combined_image/combined_image_{image_index}.png"
    combined_image.save(output_path)
```
