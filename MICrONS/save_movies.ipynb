```python
from matplotlib import pyplot as plt
import xarray as xr
from matplotlib import animation
import numpy as np
from IPython.display import HTML
import numpy as np
import cv2

ds = xr.open_dataset(
    "./function.nc",  # path to the netCDF file
    engine="netcdf4",
)  # netCDF4 file format is implemented on top of HDF5, so function.nc is also readable by any HDF5 client
```

```python
movie = ds['natural_movie'].values
```

```python
fps = 30  # 帧率
height, width = movie.shape[1], movie.shape[2]

n_segments = len(movie) // 300

for i in range(n_segments):
    start_frame = i * 300
    end_frame = (i + 1) * 300
    segment = movie[start_frame:end_frame]

    # 创建 VideoWriter
    output_filename = f"/media/ubuntu/sda/MICrONS/movies/segment_{i}.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  
    video_writer = cv2.VideoWriter(output_filename, fourcc, fps, (width, height), isColor=False)

    for frame in segment:
        frame_uint8 = frame.astype(np.uint8)
        video_writer.write(frame_uint8)

    video_writer.release()
    print(f"Saved: {output_filename}")
```

```python
import cv2
cap = cv2.VideoCapture('/home/ubuntu/Desktop/20250711beatingheart2.mp4')
```

```python
import cv2
import matplotlib.pyplot as plt

# 创建视频捕获对象 - 替换为你的视频路径

if not cap.isOpened():
    print("无法打开视频文件")
    exit()

# 读取第一帧
ret, first_frame = cap.read()

if ret:
    # 将BGR格式转换为RGB格式（Matplotlib使用RGB）
    first_frame_rgb = cv2.cvtColor(first_frame, cv2.COLOR_BGR2RGB)
    
    # 使用Matplotlib显示图像
    plt.figure(figsize=(10, 6))
    plt.imshow(first_frame_rgb)
    plt.title("视频第一帧")
    plt.axis('off')  # 关闭坐标轴
    plt.show()
else:
    print("无法读取视频帧")

# 释放资源
cap.release()
```
